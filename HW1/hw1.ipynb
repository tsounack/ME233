# Problem 1. Regression

```python
import pandas as pd

df = pd.read_excel("DataBrain.xlsx", header=[0, 1, 2, 3])
```

```python
df = df.drop([0] + list(range(len(df)-4, len(df))))
df = df.dropna(axis=1)
df
```

```python
df_cr = df.iloc[:, 6:12]
```

```python
df_cr
```

### 1.1 

```python
df_cxr_reg = df_cr.copy()
columns = [
    ('Budday17', 'CR-ten', 'epsilon', '[-]'),
    ('Budday17', 'CR-ten', 'sigma', '[kPa]'),
    ('Budday17', 'CR-com', 'epsilon', '[-]'),
    ('Budday17', 'CR-com', 'sigma', '[kPa]'),
    ('Budday17', 'CR-shr', 'gamma', '[-]'),
    ('Budday17', 'CR-shr', 'tau', '[kPa]')
]
df_cxr_reg.columns=pd.MultiIndex.from_tuples(columns)
df_cxr_reg.iloc[:, [0, 2]] = df_cxr_reg.iloc[:, [0, 2]] - 1
df_cxr_reg
```

### 1.2

```python
from sklearn.linear_model import LinearRegression

tension = LinearRegression()
tension.fit(df_cxr_reg.iloc[:, [0]], df_cxr_reg.iloc[:, [1]])

compression = LinearRegression()
compression.fit(df_cxr_reg.iloc[:, [2]], df_cxr_reg.iloc[:, [3]])

shear = LinearRegression()
shear.fit(df_cxr_reg.iloc[:, [4]], df_cxr_reg.iloc[:, [5]])

Eten = tension.coef_[0][0]
Ecom = compression.coef_[0][0]
mu = shear.coef_[0][0]

print(f"Eten = {Eten:.3f}")
print(f"Ecom = {Ecom:.3f}")
print(f"mu = {mu:.3f}")
```

### 1.3

```python
nu = 0.5
Eshr = 3 * mu * (1 + nu)
print(f"Eshr = {Eshr:.3f}")
```

### 1.4

```python
import numpy as np

mean_elastic_modulus = np.mean([Eten, Ecom, Eshr])
print(f"mean_elastic_modulus = {mean_elastic_modulus:.3f}")
```

### 1.5

```python
tension_r2 = tension.score(df_cxr_reg.iloc[:, [0]], df_cxr_reg.iloc[:, [1]])
compression_r2 = compression.score(df_cxr_reg.iloc[:, [2]], df_cxr_reg.iloc[:, [3]])
shear_r2 = shear.score(df_cxr_reg.iloc[:, [4]], df_cxr_reg.iloc[:, [5]])

print(f"R2 score for tension: {tension_r2:.3f}")
print(f"R2 score for compression: {compression_r2:.3f}")
print(f"R2 score for shear: {shear_r2:.3f}")
```

```python
from sklearn.metrics import mean_squared_error

tension_mse = mean_squared_error(df_cxr_reg.iloc[:, [1]], tension.predict(df_cxr_reg.iloc[:, [0]]))
compression_mse = mean_squared_error(df_cxr_reg.iloc[:, [3]], compression.predict(df_cxr_reg.iloc[:, [2]]))
shear_mse = mean_squared_error(df_cxr_reg.iloc[:, [5]], shear.predict(df_cxr_reg.iloc[:, [4]]))

print(f"MSE for tension: {tension_mse:.6f}")
print(f"MSE for compression: {compression_mse:.6f}")
print(f"MSE for shear: {shear_mse:.6f}")
```

### 1.6

We can see that the results for tension seem very good: we observe a 0.988 R2 score, and a 2.4*10^-5 MSE score. For both metrics, this is the best out of the three experiments.

Interestingly, we notice that while compression has a higher R2 score than shear (0.93 vs 0.92), it has a higher MSE score (3.7*10^-3 vs 4.6*10^-4). This means that the model for compresion is better at capturing the overall trend of the data, while the model for shear makes predictions that are on average closer to the actual values.
